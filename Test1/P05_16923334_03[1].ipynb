```python
print("1.  :")
main_table["property_tax"].plot(kind="hist",bins =[0 ,20 ,40 ,60 ,80 ,100] , rwidth =0.8)
# plotting histogram on property tax with 20 as its bins

print("-------------------------------------------------------")
```

```python
print("2.  :")
# selecting the data with area below 10000
sort2 = main_table.loc[(main_table["area"] < 10000)]
# converting and plotting to the pie chart
sort2["city"].value_counts().plot(kind = "pie")

print("-------------------------------------------------------")
```

```python
print("3.  :")
# selecting only the data which is furnished
sort3 = main_table.loc[(main_table["furniture"] == "furnished")]
# plotting thr data scatter graph based on insurance and rent amount
main_table.plot(kind="scatter", x="fire_insurance", y="rent_amount")

print("-------------------------------------------------------")
```

```python
print("4.  :")
# sorting the data wih parking spaces equal or below to 7 as maximum
sort4 = main_table.loc[(main_table["parking_spaces"] <= 7)]
# plotting the horizontal bar graph based on the data
sort4["city"].value_counts().plot(kind = "barh")

print("-------------------------------------------------------")
```

```python
print("5.  :")
import pandas as pd # Importing pandas package to be used
import matplotlib.pyplot as plt # Importing matplotlib package to be used

main_table = pd.read_csv("P05-02.csv")
# selecting the data with specific specifed city
sort5 = main_table.loc[(main_table["city"] == "Rio de Janeiro") | (main_table["city"] == "Sao Paulo") | (main_table["city"] == "Porto Alegre")]
# plotting a quite complicated stacked bar graph showing some properties like animal and from which city
sort5.groupby(["animal","city"]).size().unstack().plot(kind="bar",stacked=True)

print("-------------------------------------------------------")
```
