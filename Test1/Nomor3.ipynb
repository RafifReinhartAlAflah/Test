```python
import pandas as pd 
df = pd.read_csv("P05-02.csv")
print ("1. ")
df[["property_tax"]].plot(kind="hist",bins=[0,20,40,60,80,100], rwidth= 0.8, ylabel="property_tax")
```

```python
print("2. ")
data = df.loc[(df["area"] <1000)]
data["city"].value_counts().plot(kind="pie", ylabel="Area")
```

```python
print("3. ")
df.plot(kind="scatter",x ="fire_insurance", y="rent_amount")
```

```python
print("4. ")
data = df.loc[(df["parking_spaces"] <= 7)]
data["city"].value_counts().plot(kind="barh", x="city", y="parking_spaces")
```

```python
print("5. :")
data = df.loc[(df["city"] == "Rio de Janeiro") | (df["city"] == "Sao Paulo") | (df["city"] == "Porto Alegre")]
data.groupby(["animal","city"]).size().unstack().plot(kind="bar",stacked = True)
```
