```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

wavelength = np.array([367.22, 432.108, 498.1, 556.66, 703.29])
voltage = np.array([3.4, 2.88, 2.51, 2.2, 1.74])

plt.scatter(wavelength, voltage, color='purple', label='Data')

slope, intercept, r_value, p_value, std_err = stats.linregress(wavelength, voltage)
line = slope * wavelength + intercept

plt.plot(wavelength, line, color='black', label='Regresi Linear')

plt.xlabel('λmaks (nm)')
plt.ylabel('Von (V)')
plt.title('Regresi Linear Konstanta Planck')
plt.legend()

plt.grid(True)
plt.show()

print("Persamaan Regresi Linear: V_on =", slope, "* lambda_max +", intercept)
print("Nilai r^2:", r_value**2)
```
